```python
cc['t']=np.arange(1,43)
cc['Q']=['Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2','Q3','Q4','Q1','Q2']
```

```python
c=pd.get_dummies(cc['Q'])
```

```python
df=[cc,c]
data=pd.concat(df,axis=1,join='inner')
```

```python
data['t2']=np.square(data.t)
data['log_S']=np.log(data.Sales)
```

```python
data
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Q",y="Sales",data=cc)
```

## Creating different models and computing there RMSE(Root Mean Square Error) for best Forcasting model

```python
train=data.head(38)
test=data.tail(4)
```

```python
#linear model
linear=smf.ols('Sales~t',data=train).fit()
pred_lin=pd.Series(linear.predict(pd.DataFrame(test['t'])))
rmse_lin=np.sqrt((np.mean(np.array(test['Sales'])-np.array(pred_lin))**2))
rmse_lin
```

```python
#quadratic model
quad=smf.ols('Sales~t+t2',data=train).fit()
pred_quad=pd.Series(quad.predict(pd.DataFrame(test[['t','t2']])))
rmse_quad=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_quad))**2))
rmse_quad
```

```python
#exponential model
expo=smf.ols('log_S~t',data=train).fit()
pred_exp=pd.Series(expo.predict(pd.DataFrame(test['t'])))
rmse_expo=np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_exp)))**2))
rmse_expo
```

```python
#additive seasonality
additive=smf.ols('Sales~Q1+Q2+Q3+Q4',data=train).fit()
pred_add=pd.Series(additive.predict(pd.DataFrame(test[['Q1','Q2','Q3','Q4']])))
rmse_add=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add))**2))
rmse_add
```

```python
#additive seasonality with linear trend
addlinear=smf.ols('Sales~t+Q1+Q2+Q3+Q4',data=train).fit()
pred_add_linear=pd.Series(addlinear.predict(pd.DataFrame(test[['t','Q1','Q2','Q3','Q4']])))
rmse_add_linear=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_linear))**2))
rmse_add_linear
```

```python
#additive seasonality with quadratic trend
addquad=smf.ols('Sales~t+t2+Q1+Q2+Q3+Q4',data=train).fit()
pred_add_quad=pd.Series(addquad.predict(pd.DataFrame(test[['t','t2','Q1','Q2','Q3','Q4']])))
rmse_add_quad=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_quad))**2))
rmse_add_quad
```

```python
#multiplicative seasonality
multsea=smf.ols('log_S~Q1+Q2+Q3+Q4',data=train).fit()
pred_mult= pd.Series(multsea.predict(pd.DataFrame(test[['Q1','Q2','Q3','Q4']])))
rmse_mult= np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_mult)))**2))
rmse_mult
```

```python
#multiplicative seasonality with linear trend
mult_lin=smf.ols('log_S~t+Q1+Q2+Q3+Q4',data=train).fit()
pred_mult_lin= pd.Series(mult_lin.predict(pd.DataFrame(test[['t','Q1','Q2','Q3','Q4']])))
rmse_mult_lin=np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_mult_lin)))**2))
rmse_mult_lin
```

```python
#multiplicative seasonality with quadratic trend
mult_quad= smf.ols('log_S~t+t2+Q1+Q2+Q3+Q4',data=train).fit()
pred_mult_quad= pd.Series(mult_quad.predict(test[['t','t2','Q1','Q2','Q3','Q4']]))
rmse_mult_quad=np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_mult_quad)))**2))
rmse_mult_quad
```

```python
d={'Model':pd.Series(['rmse_mult_quad','rmse_add','rmse_add_linear','rmse_add_quad','rmse_expo','rmse_lin','rmse_mult','rmse_mult_lin','rmse_quad']),'RMSE_Values':pd.Series([rmse_mult_quad,rmse_add,rmse_add_linear,rmse_add_quad,rmse_expo,rmse_lin,rmse_mult,rmse_mult_lin,rmse_quad])}
table_rmse=pd.DataFrame(d)
table_rmse.sort_values(['RMSE_Values'])
```
